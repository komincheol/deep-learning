```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

n = 1000    # number of points to create
xcord, ycord = np.zeros(n), np.zeros(n)
markers, colors = [], []

fw = open('testSet.txt', 'w')

for i in range(n):
    [r0, r1] = np.random.standard_normal(2)
    myClass = np.random.uniform(0, 1)

    if myClass <= 0.16:
        fFlyer = np.random.uniform(22000, 60000)
        tats = 3 + 1.6 * r1
        markers.append(20)
        colors.append('red')
        classLabel = 1    # 'didntLike'
        print("%d, %f, class1" % (fFlyer, tats))
    elif myClass <= 0.33:
        fFlyer = 6000 * r0 + 70000
        tats = 10 + 3 * r1 + 2 * r0
        markers.append(20)
        colors.append('red')
        classLabel = 1    # 'didntLike'
        print("%d, %f, class1" % (fFlyer, tats))
    elif myClass <= 0.66:
        fFlyer = 5000 * r0 + 10000
        tats = 3 + 2.8 * r1
        markers.append(30)
        colors.append('green')
        classLabel = 2    # 'smallDoses'
        print("%d, %f, class2" % (fFlyer, tats))
    else:
        fFlyer = 10000 * r0 + 35000
        tats = 10 + 2.0 * r1
        markers.append(50)
        colors.append('blue')
        classLabel = 3    # 'largeDoses'
        print("%d, %f, class3" % (fFlyer, tats))

    if tats < 0: tats = 0
    if fFlyer < 0: fFlyer = 0
    xcord[i], ycord[i] = fFlyer, tats;

    fw.write("%d\t%f\t%f\t%d\n" % (fFlyer, tats, np.random.uniform(0.0, 1.7), classLabel))

fw.close()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(xcord, ycord, c=colors, s=markers)
type1 = ax.scatter([-10], [-10], s=20, c='red')
type2 = ax.scatter([-10], [-15], s=30, c='green')
type3 = ax.scatter([-10], [-20], s=50, c='blue')
ax.legend([type1, type2, type3], ["Class 1", "Class 2", "Class 3"], loc=2)
ax.axis([-5000, 100000, -2, 25])
plt.xlabel('Frequent Flyer Miles Earned Per Year')
plt.ylabel('Percentage of Body Covered By Tatoos')
plt.show()
```
